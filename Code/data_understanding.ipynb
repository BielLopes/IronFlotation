```python
from statistics import mean, mode


k = 0
per_step_samples = []
i = 0
data_old = data_set['date'][26814]

one_hour_step = pd.Timedelta(3.6e12)
one_hour_steps = []

for w, date in enumerate(data_set['date'].loc[26814:]):
    if date != data_old:
        # if i == 179:
        one_hour_steps.append(date - data_old == one_hour_step)

        if date - data_old != one_hour_step:
            print(date)
            print(data_old)
            print(date - data_old)
            print(w)
        
        if i != 179:
            print(f"{k} - {date} - {i}")

        data_old = date
        k += 1

        per_step_samples.append(i)
        i = 0
    else:
        i += 1

print(k)

data_mode = mode(per_step_samples)
complete_data = [sizes == data_mode for sizes in per_step_samples]
print(sum(complete_data))

print(sum(one_hour_steps))
```

```python
import math

def round_up_float_to_2_decimals(num):
    return math.ceil(num * 100) / 100

util_data_set = data_set['Flotation Column 03 Air Flow'].loc[26814:].copy()
h = 0
all_data = []
for data in util_data_set:
    if h > 178:
        break
    
    all_data.append(data)
    # print(type(data))
    h += 1

print(round_up_float_to_2_decimals(mean(all_data)))
```
